```python
df_places = df[['City','State','Region','Sales']]
```

```python
grouped_data = df_places.groupby(['Region'], as_index=False).sum()
grouped_data.sort_values(by='Sales', ascending=False, inplace=True)

plt.figure(figsize=(10,5))
plt.bar(grouped_data['Region'], grouped_data['Sales'], align='center',)
plt.title("Sales Generated by Region")
plt.xlabel("Region")
plt.ylabel("Sales")


plt.show()
```

```python
grouped_data = df_places.groupby(['State'], as_index=False).sum()
grouped_data.sort_values(by='Sales', ascending=False, inplace=True)

plt.figure(figsize=(16,8))
plt.bar(grouped_data['State'], grouped_data['Sales'], align='center',)
plt.xlabel("State")
plt.ylabel("Sales")
plt.title("Sales Generated by State")
plt.xticks(rotation=90)

plt.show()
```

```python
top_cities = df.groupby("City")["Sales"].sum().nlargest(10)

plt.figure(figsize=(8, 6))

top_cities.plot(kind="bar", color="red")
plt.xlabel("City")
plt.ylabel("Sales Amount")
plt.title("Top 10 Cities with Highest Sales", color="Blue")
plt.xticks(rotation=45)

plt.show()
```

```python
df_relations = df[['Ship_Mode','Sales']]
```

```python
Ship_Mode_Sales_Profit = df_relations.groupby(['Ship_Mode']).sum().reset_index(False)
Ship_Mode_Sales_Profit =Ship_Mode_Sales_Profit.sort_values('Sales', ascending=False)

plt.figure(figsize=(10,15))
plt.subplot(3,1,1)
plt.bar(x=Ship_Mode_Sales_Profit['Ship_Mode'], height=Ship_Mode_Sales_Profit['Sales'], color='Green')
plt.title("relationship between ship mode and sales", size=20, pad=20)


plt.show()
```

```python

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

df['Year'] = df['Order_Date'].dt.year

target_year = 2017

df_target_year = df[df['Year'] == target_year]

product_sales = df_target_year.groupby('Product_Name')['Sales'].sum().reset_index()

top_selling_products = product_sales.sort_values(by='Sales', ascending=False)

print("Top Selling Products in", target_year)
print(top_selling_products)

plt.figure(figsize=(10, 6))
plt.bar(top_selling_products['Product_Name'][:10], top_selling_products['Sales'][:10], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title(f'Top 10 Selling Products in {target_year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

df['Year'] = df['Order_Date'].dt.year

target_year = 2017

df_target_year = df[df['Year'] == target_year]

product_sales = df_target_year.groupby('Product_Name')['Sales'].sum().reset_index()

least_selling_products = product_sales.sort_values(by='Sales', ascending=True)

print("Least Selling Products in", target_year)
print(least_selling_products.head(10))

plt.figure(figsize=(10, 6))
plt.bar(least_selling_products['Product_Name'][:10], least_selling_products['Sales'][:10], color='lightcoral')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title(f'Least 10 Selling Products in {target_year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

customer_products = df.groupby('Customer_Name')['Product_Name'].unique().reset_index()

print("Customers and their Bought Products:")
for index, row in customer_products.sample(5).iterrows():
    customer_name = row['Customer_Name']
    bought_products = row['Product_Name']
    print(f"{customer_name} bought the following products:")
    for product in bought_products:
        print(f"- {product}")
    print()
```

```python
customer_product_count = df.groupby('Customer_Name')['Product_Name'].nunique().reset_index()

top_buying_customers = customer_product_count.sort_values(by='Product_Name', ascending=False)

most_products_customer = top_buying_customers.iloc[0]['Customer_Name']
print(f"The customer who bought the most products is: {most_products_customer}")

plt.figure(figsize=(10, 6))
plt.bar(top_buying_customers['Customer_Name'][:10], top_buying_customers['Product_Name'][:10], color='green')
plt.xlabel('Customer Name')
plt.ylabel('Number of Products Bought')
plt.title('Top 10 Customers Buying the Most Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
customer_order_count = df.groupby('Customer_Name')['Order_ID'].nunique().reset_index()

top_regular_customers = customer_order_count.sort_values(by='Order_ID', ascending=False)

print("Top Regular Customers:")
print(top_regular_customers.head(10))

plt.figure(figsize=(10, 6))
plt.bar(top_regular_customers['Customer_Name'][:10], top_regular_customers['Order_ID'][:10], color='orange')
plt.xlabel('Customer Name')
plt.ylabel('Number of Orders Placed')
plt.title('Top 10 Regular Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

df['Year'] = df['Order_Date'].dt.year

customer_product_count_yearly = df.groupby(['Year', 'Customer_Name'])['Product_Name'].nunique().reset_index()

top_customers_yearly = customer_product_count_yearly.loc[customer_product_count_yearly.groupby('Year')['Product_Name'].idxmax()]

print("Top Customers Year-wise:")
print(top_customers_yearly)

for year in df['Year'].unique():
    top_customers_year = top_customers_yearly[top_customers_yearly['Year'] == year]
    plt.figure(figsize=(10, 6))
    plt.bar(top_customers_year['Customer_Name'], top_customers_year['Product_Name'], color='purple')
    plt.xlabel('Customer Name')
    plt.ylabel('Number of Products Bought')
    plt.title(f'Top Customer(s) in {year} Who Bought the Most Products')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
customer_product_count_region = df.groupby(['Region', 'Customer_Name'])['Product_Name'].nunique().reset_index()

top_customers_region = customer_product_count_region.loc[customer_product_count_region.groupby('Region')['Product_Name'].idxmax()]

print("Top Customers Region-wise:")
print(top_customers_region)

for region in df['Region'].unique():
    top_customers_region_year = top_customers_region[top_customers_region['Region'] == region]
    plt.figure(figsize=(10, 6))
    plt.bar(top_customers_region_year['Customer_Name'], top_customers_region_year['Product_Name'], color='brown')
    plt.xlabel('Customer Name')
    plt.ylabel('Number of Products Bought')
    plt.title(f'Top Customer(s) in {region} Region Who Bought the Most Products')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

df['Year'] = df['Order_Date'].dt.year.astype(int)

yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

max_sales_year = yearly_sales.loc[yearly_sales['Sales'].idxmax()]

print("Year with the Highest Sales:")
print(max_sales_year)

plt.figure(figsize=(10, 6))
plt.bar(yearly_sales['Year'], yearly_sales['Sales'], color='blue')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales in Each Year')
plt.xticks(yearly_sales['Year'].astype(int), rotation=45, ha='right')
plt.show()
```
